## For more pages

```python
def pages_to_scrape(min_page, max_page):
    pages = []
    url = "https://www.propertypro.ng/property-for-rent?page=" 
    for i in range(min_page, max_page+1):
            pages.append(url + str(i))
    return pages
```

```python
#pages
```

```python
def apart_listing(pages):

    Description = []
    Links = []
    Locations = []
    State = []
    Price = []
    Bedrooms = []
    Bathrooms = []
    Toilets = []
    Agent_number = []
    Added_on = []
    Updated_on = []
    Area = []

    for page in pages:
        try:
                 response = requests.get(page)
        except:
                break
        data_ = response.text
        soup = BeautifulSoup(data_, "html.parser")
        listings = soup.find_all(class_ = "listings-property-title2")
        for listing in listings:
            Description.append(listing.text)
        locations = soup.select("div.single-room-text > h4")
        Locations_ = [location.text for location in locations]
        for location in locations:
            Locations.append(location.text)
        links = soup.select("div.single-room-text > a")
        for link in links:
            Links.append("https://www.propertypro.ng" + link["href"])

        for loc in Locations_:
            State.append(loc.split()[-1])
        for loc in Locations_:
            Area.append(loc.split()[-2])

        price = soup.select("div.n50 > h3")
        for pric in price:
            Price.append(pric.text.split()[-1])
        features = soup.select("div.fur-areea > span")
        for bed in features:
            if bed.text.split()[-1]=="beds":
                Bedrooms.append(bed.text)
        for baths in features:
            if baths.text.split()[-1] == "baths":
                Bathrooms.append(baths.text)
        for toilets in features:
            if toilets.text.split()[-1] == "Toilets":
                Toilets.append(toilets.text)
        phones = soup.select("div.phone-icon")
        for agent in phones:
            Agent_number.append(agent.text.strip())
        dates = soup.select("div.single-room-text > h5")
        Date_added = [date.text for date in dates]
        for date in Date_added:
            Added_on.append(date.partition("Added")[2].strip())
        for date in Date_added:
            if "Update" in date:
                Updated_on.append(date.partition("Updated")[2].partition(", Added")[0].strip())
            else:
                Updated_on.append(date.partition("Added")[2].strip())
    df = pd.DataFrame({'Description':Description, 'Location': Locations, 'Links': Links, 'State': State, 
                      'Area':Area, 'Price': Price,
                      'Bedrooms': Bedrooms, 'Bathrooms': Bathrooms, 'Toilets': Toilets,
                      'Agent_number': Agent_number, 'Added_on': Added_on, 'Updated_on': Updated_on})
    df.to_csv("apartment_rent_"+str(len(pages))+".csv", index=False)
    return df

```
